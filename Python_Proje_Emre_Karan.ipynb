## Proje Soruları

1. Nobel Ödüllerini en çok kazanan ilk on ülkeyi bulunuz.
2. Nobel Ödüllerini kazanan ilk kadınları listeleyiniz.
3. Nobel Ödüllerini kazanan ilk erkekleri listeleyiniz.
4. Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?

6. 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.
7. 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)
8. Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

# Nobel Ödüllerini en çok kazanan ilk on ülkeyi bulunuz.

```python
csv_dosyasi = "proje.csv"
veri = pd.read_csv(csv_dosyasi)
ulke_toplam = veri['birth_country'].value_counts().reset_index()
ulke_toplam = veri['birth_country'].value_counts().reset_index()
ulke_toplam.columns = ['Ülke', 'Nobel Ödülü Sayısı']
sirali_ulke_toplam = ulke_toplam.sort_values(by='Nobel Ödülü Sayısı', ascending=False)
ilk_10_ulke = sirali_ulke_toplam.head(10)
print("Nobel Ödüllerini En Çok Kazanan İlk 10 Ülke:")
print(ilk_10_ulke)
```

# Nobel Ödüllerini kazanan ilk kadınları listeleyiniz.

```python
female_winners = veri[veri["sex"] == "Female"]
female_winners = female_winners[["year", "category", "full_name"]]
print("Nobel Ödülünü Kazanan İlk Kadınlar:\n", female_winners.head(10))
```

# Nobel Ödüllerini kazanan ilk erkekleri listeleyiniz.

```python
male_winners = veri[veri["sex"] == "Male"]
male_winners = male_winners[["year", "category", "full_name"]]
print("Nobel Ödülünü Kazanan İlk Erkekler:\n", male_winners.head(10))
```

# Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?

```python
import matplotlib.pyplot as plt
ulke_odul_sayisi = veri["birth_country"].value_counts()
en_fazla_odul_alan_ulke = ulke_odul_sayisi.idxmax()
en_fazla_odul_alan_ulkenin_yillara_gore_odulleri = veri[veri["birth_country"] == en_fazla_odul_alan_ulke]["year"].value_counts().sort_index()
hakimiyet_baslangic_yili = en_fazla_odul_alan_ulkenin_yillara_gore_odulleri.idxmax()
plt.figure(figsize=(12, 6))
plt.plot(en_fazla_odul_alan_ulkenin_yillara_gore_odulleri.index, en_fazla_odul_alan_ulkenin_yillara_gore_odulleri.values, marker='o', linestyle='-')
plt.title(f"{en_fazla_odul_alan_ulke}'nin Nobel Ödülleri Yıllara Göre Dağılımı")
plt.xlabel("Yıl")
plt.ylabel("Ödül Sayısı")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
print(f"{en_fazla_odul_alan_ulke} en fazla ödül aldığı yıldan sonra hakimiyetini {hakimiyet_baslangic_yili} yılından itibaren sağlamıştır.")
```

# Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
nobel_data = pd.read_csv("proje.csv")
nobel_data['Age'] = pd.to_datetime(nobel_data['death_date']).dt.year - pd.to_datetime(nobel_data['birth_date']).dt.year
plt.figure(figsize=(10, 6))
sns.histplot(nobel_data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Nobel Ödülü Kazananların Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='category', data=nobel_data, palette='pastel')
plt.title('Nobel Ödülü Kazananların Kategorilere Göre Dağılımı')
plt.xlabel('Kazanan Sayısı')
plt.ylabel('Ödül Kategorisi')
plt.show()
```

```python
gender_counts = nobel_data['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('Nobel Ödülü Kazananların Cinsiyet Dağılımı')
plt.show()
```

```python
nobel_data['Year_Group'] = pd.cut(nobel_data['year'], bins=range(1900, 2025, 5), right=False)
year_counts = nobel_data['Year_Group'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('Nobel Ödülü Kazananların Yıllara Göre Dağılımı')
plt.xlabel('Yıl Aralığı')
plt.ylabel('Kazanan Sayısı')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
veri = pd.read_csv('proje.csv')
fizik_verileri = veri[(veri['category'] == 'Physics') & (veri['year'] > 2000)].copy()
fizik_verileri['birth_date'] = pd.to_datetime(fizik_verileri['birth_date'])
fizik_verileri['death_date'] = pd.to_datetime(fizik_verileri['death_date'])
fizik_verileri['Yaş'] = (fizik_verileri['death_date'] - fizik_verileri['birth_date']).dt.days // 365
fizik_verileri = fizik_verileri[['birth_country', 'Yaş']]
plt.figure(figsize=(10,12))
sns.barplot(x='Yaş', y='birth_country', data=fizik_verileri, color='skyblue')
plt.xlabel('Yaş')
plt.ylabel('Doğum Ülkesi')
plt.title('Fizik Kategorisinde 2000 yılından Sonra Kazananların Ülkeleri ve Yaşları')
plt.grid(axis='x')
plt.show()
```
